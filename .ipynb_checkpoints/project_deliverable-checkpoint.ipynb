**Description**: This video discusses the debate over whether Tylenol, or maternal acetaminophen use, causes autism, and whether scientific researchers failed to reveal the truth to the public.

Drawing on her own research background and medical degree, the video creator argues that although numerous studies have investigated the relationship between acetaminophen and autism, no clear evidence has emerged, and thus no causal claim can be made. In contrast, the Trump administration claimed to be the first to identify a causal link between maternal Tylenol use and autism, implying to the public that the scientific community has not fulfilled its responsibility to society.

## Research question:
To what extend does the public believe in the relation between maternal acetaminophen use and autism and the work scientists do in researching autism?

After re-evaluating the content of the video, we recognized that the debate is not limited to biomedical causality. It also concerns the political framing of scientific responsibility. Because the video creator positions herself with scientific researchers, comments that support her were coded as supporting the scientific side.

## Prompting stategy reflection

At the beginning, we wrote the prompt based on what we thought the two concepts meant in a more theoretical sense. After we compared the model outputs with the human-coded data, the inter-rater reliability was much lower than we expected. This showed us that the LLM was not following the same coding patterns as the human coders, so we needed to adjust our approach.
The main change we made was shifting from a theory-based definition to a data-driven one. Instead of assuming how people should code the comments, we studied the CSV file closely and identified how human coders were actually making their decisions. For example, for Concept 2, we originally thought the comment needed to explicitly support scientific reasoning to get a 1. But in the real data, even simple praise toward the video or any anti-Trump expression counted as support for the author and the scientific community. Once we noticed this, we rewrote the prompt so the LLM would follow the same pattern.
We also simplified the rules and added clear boundaries using examples taken directly from the human-coded cases. This helped the model imitate human behavior rather than overinterpret the comments. In other words, instead of trying to understand the issue at a deeper level, the LLM now focuses on matching human decisions.
Overall, refining the prompt so it better reflected the actual human coding patterns helped us improve agreement between the human coders and the model.

## Results

To answer our research question of how strongly the public believes in a causal relationship between maternal acetaminophen use and autism, and how much they trust the scientists studying autism, we analyzed coded TikTok comments from the selected video. Across the dataset, comments endorsing the Tylenolâ€“autism causal claim (Concept 1) and comments rejecting causation or supporting scientific evidence (Concept 2) appeared at relatively similar frequencies. This suggests that the comment section reflects a divided public, with a substantial portion of users repeating or agreeing with unsupported causal claims, while another large group actively pushes back by emphasizing lack of evidence, praising the video creatorâ€™s scientific explanation, or expressing trust in researchers. The presence of strong Concept 2 comments indicates that many users do understand the difference between correlation and causation and recognize the scientific communityâ€™s efforts to research autism responsibly.

Engagement patterns give additional insight into how these beliefs circulate. As shown in the charts, Concept 2 comments tended to receive more likes, meaning that the broader viewing audience reacted more positively to scientifically aligned explanations. In contrast, Concept 1 comments received more replies, suggesting that misinformation-leaning or politically charged claims generated more debate and interaction. Because TikTok ranks comments partly based on replies, this structure may unintentionally elevate comments endorsing the causal link, even though they are not the most positively received. This dynamic complicates the overall picture: while many commenters reject the Tylenolâ€“autism claim and support scientific methods, the platform may amplify the appearance of disagreement or controversy.

The age distribution adds another layer to understanding public belief. The age graph shows that Concept 2 comments come from slightly older users on average, while Concept 1 and neutral comments appear more often among younger users. This means older commenters seem more aligned with scientific reasoning and less likely to accept unsupported causal claims. Younger users were more mixed, with some rejecting the claim and others repeating it. Together, these results show that public belief is not one-sided. Many users openly support scientific evidence and trust autism research, but misinformation can still gain attention because of platform dynamics and the commenting behaviors of younger audiences.
